# croping images

```python
lake_bgr=cv2.imread("/kaggle/input/boat1q/New_Zealand_Boat.jpg",cv2.IMREAD_COLOR)
lake_rgb=lake_bgr[:,:,::-1]
```

```python
plt.imshow(lake_rgb)
```

```python
boat=lake_rgb[200:400,300:600]
plt.imshow(boat)
```

# resizing images

```python
boat_2wice=cv2.resize(boat,None,fx=2,fy=2)
plt.imshow(boat_2wice)
```

# image fliping

```python
# lets fliping along the horizontal axis
hor_flip=cv2.flip(lake_rgb,1)
vert_flip=cv2.flip(lake_rgb,0)
both_flip=cv2.flip(lake_rgb,-1);
#show images
plt.figure(figsize=[50,30])
plt.subplot(141);plt.imshow(hor_flip);plt.title("horizontal-flip")
plt.subplot(142);plt.imshow(hor_flip);plt.title("vertical flip")
plt.subplot(143);plt.imshow(hor_flip);plt.title("both sides flip")
plt.subplot(144);plt.imshow(lake_rgb);plt.title("original")
```
